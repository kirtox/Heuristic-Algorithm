```python
import time
import random
import numpy as np

def solve(vlist,wlist,totalweight,totalLength):
    tmp = []
    init_weight = totalweight
    Final_value = 0
    Final_weight = 0
    #resArr = np.zeros((totalLength+1,weight+1),dtype=int)
    #print(resArr)
    #print(len(resArr))
    for i in range(1,(totalLength+1)+1): #1-7
        #for j in range(1,weight+1):.
        value = 0
        notFull = 0
        weight = init_weight
        while(notFull == 0):
            index = random.randint(0,totalLength)
            print('index: ',index)
            if weight-wlist[index] > 0:
                print('test1')
                weight -= wlist[index]
                value += vlist[index] 
                print('weight: ',weight)
                print('value: ',value,'\n')
            elif weight-wlist[index] == 0:
                print('test2')
                weight -= wlist[index]
                value += vlist[index]
                notFull = 1
                print('weight: ',weight)
                print('value: ',value,'\n')
            else:
                print('test3')
                notFull = 1
                print('weight: ',weight)
                print('value: ',value,'\n')

            if Final_value < value:
                Final_value = value
                Final_weight = init_weight - weight
            
        print('#####################################\n')
    #print(resArr)
    
    return Final_weight, Final_value

start = time.time() 
v = [442,525,511,593,546,564,617]
w = [41,50,49,59,55,57,60] #0-6
weight = 100
n = 6
result_w, result_v = solve(v,w,weight,n)
end = time.time()

print('Final_Value: ',result_v,'\n')
print('Final_Weight: ',result_w,'\n')
print('Period= ', round(end-start, 8), ' sec\n')
```
